```python
from pandas import DataFrame

def data_normalization(df: DataFrame, days_range: int=24, include_cur_row: bool=False):
    '''
    Return the normalized data:
    + df: the dataframe to be normalized
    + range: the number of previous rows (or including the current row) to be considered in the normalization
    + include_cur_row: True if we consider the current row in the normalization process (calculate mean and std
    using the current row and (range-1) previous rows), False if we want to use all the passed data for normalization 
    processing ((calculate mean and std using (range) previous rows))
    '''
    
    df_roll = None

    if include_cur_row == False:
        df_roll = df.rolling(days_range, closed='left')
    else:
        df_roll = df.rolling(days_range)
    res_df = (df - df_roll.mean()) / df_roll.std()
    res_df.replace([np.inf, -np.inf, np.nan], 0, inplace=True)
    return res_df[days_range:] #
```

### Profit evaluation

```python
stop_loss_threshold = 0.1

def calculate_profit(df_test, i):
    global stop_loss_threshold
    
    prev_x = df_test.iloc[i-1]
    x = df_test.iloc[i]
    
    if x['pred'] == 0: # and prev_x['percen_change'] < 0:
        if x['high_diff_open'] < stop_loss_threshold:
            return -x['percen_change']
        return -stop_loss_threshold
    if x['pred'] == 1: # and prev_x['percen_change'] > 0:
        if x['low_diff_open'] > -stop_loss_threshold:
            return x['percen_change']
        return -stop_loss_threshold
    return 0
```

```python
profit_list = []

for i in range(1, df_test.shape[0]):
    profit = calculate_profit(df_test, i)
    if profit == 0:
        df_test.at[i,'pred'] = -1
    profit_list.append(profit)
```

```python
fee_rate = 0.00012*2
```

**Calculate the number of decisions**

```python
num_of_decisions = (df_test['pred'] != -1).sum()
num_of_no_orders = ((df_test['pred'] == df_test['pred'].shift(1)) & (df_test['pred'] != -1)).sum()
num_of_decisions = num_of_decisions - num_of_no_orders
```

```python
print("Number of decisions:", num_of_decisions)
```

**Calculate the fee price**

```python
fee_price = num_of_decisions*fee_rate
```

```python
print("Number of decisions:", num_of_decisions)
print("Fee price:", fee_price)
```

```python
print("Profit without trading fee:", sum(profit_list))
print("Profit with fee included:", sum(profit_list) - fee_price)
```

**Cumulative Profit**

```python
k = np.array(((df_test['pred'] == df_test['pred'].shift(1)) & (df_test['pred'] != -1)))
```

```python
balance = 1
i = 0

while i < len(profit_list):
    sum_p = fee_rate
    if profit_list[i] != 0:
        sum_p = profit_list[i]
        while i < len(profit_list)-1 and k[i+1] == True:
            sum_p += profit_list[i+1]
            i += 1
    balance *= (1 + sum_p - fee_rate)
    i += 1
    
print(balance)
```
