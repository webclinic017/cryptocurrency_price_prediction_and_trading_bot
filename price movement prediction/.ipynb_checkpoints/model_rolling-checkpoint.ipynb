# Data preparation

### Modules

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import sys
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.metrics import roc_auc_score, mean_squared_error, mean_absolute_percentage_error, classification_report, accuracy_score
import math 
from sklearn.linear_model import Ridge, LogisticRegression, Lasso
```

```python
from pandas import DataFrame

def data_normalization(df: DataFrame, days_range: int=24, include_cur_row: bool=False):
    '''
    Return the normalized data:
    + df: the dataframe to be normalized
    + range: the number of previous rows (or including the current row) to be considered in the normalization
    + include_cur_row: True if we consider the current row in the normalization process (calculate mean and std
    using the current row and (range-1) previous rows), False if we want to use all the passed data for normalization 
    processing ((calculate mean and std using (range) previous rows))
    '''
    
    df_roll = None

    if include_cur_row == False:
        df_roll = df.rolling(days_range, closed='left')
    else:
        df_roll = df.rolling(days_range)
    res_df = (df - df_roll.mean()) / df_roll.std()
    res_df.replace([np.inf, -np.inf, np.nan], 0, inplace=True)
    return res_df[days_range:] #
```

### Data loading

```python
# df is the dataframe in which the data was already normalized
df = pd.read_csv('data/normalized_data.csv')
df.drop(columns=["Unnamed: 0"], inplace=True)

# ori_df is the original dataframe in which the data is not normalized yet
ori_df = pd.read_csv('data/features_with_cryptoquant_data_df.csv')
ori_df.drop(columns=["Unnamed: 0"], inplace=True)
ori_df.reset_index(drop=True, inplace=True)
ori_df = ori_df[-df.shape[0]:]
ori_df.reset_index(drop=True, inplace=True)
```

**Train val test split**

```python
X, y = df.drop(columns=['label']), df['label'][:]
X_train, X_valid, X_test = X[:int(X.shape[0]*0.7)], X[int(X.shape[0]*0.7):int(X.shape[0]*0.85)], X[int(X.shape[0]*0.85):]
y_train, y_valid, y_test = y[:int(X.shape[0]*0.7)], y[int(X.shape[0]*0.7):int(X.shape[0]*0.85)], y[int(X.shape[0]*0.85):]
```

# Training

```python
import lightgbm as lgb
from lightgbm import LGBMClassifier
train_data = lgb.Dataset(X_train, y_train)
valid_data = lgb.Dataset(X_valid, y_valid, reference=train_data)
```

```python
depth = 4
rate = 0.2
lr = 0.002

params = {'objective': 'binary' ,
                      'metric': ['cross_entropy'],
                      'boosting':'gbdt', 
                      'num_boost_round':4096,
                      'early_stopping_round':32,
                      'learning_rate': lr,
                      'max_depth': depth,
                      'num_leaves': 2**depth,
                      'feature_fraction': rate,
                      'verbosity': 1}

print(params)
```

```python
model = lgb.train(params, train_data,                     
                  valid_sets=[train_data, valid_data],
                  valid_names=['train', 'valid'])
```

```python
threshold = max(l)
delta = 0.05

def decision_mapping(x):
    global threshold, delta
    
    if x >= threshold + delta: # Long decision
        return 1 
    elif x < threshold - delta: # Short decision
        return 0 
    return -1 # No decision

df_test['pred'] = np.vectorize(decision_mapping)(model.predict(X_test))
```

### Interest evaluation

```python
threshold = 0.005

def calculate_interest(percen_change, high_diff_open, low_diff_open, pred):
    global threshold
    if pred == 0:
        if percen_change >= 0:
            return max(-percen_change, -threshold)
        return -percen_change
    if pred == 1:
        if percen_change < 0:
            return max(percen_change, -threshold)
        return percen_change
    return 0
```

```python
interest_list = []
for i in range(df_test.shape[0]):
    x = df_test.iloc[i]
    interest = calculate_interest(x['percen_change'], x['high_diff_open'], x['low_diff_open'], x['pred'])
    interest_list.append(interest)
```

```python
fee_rate = 0.00012+0.00030
num_of_decisions = (np.array(interest_list) != 0).sum()
fee_price = num_of_decisions*fee_rate

print("Number of decisions:", (np.array(interest_list) != 0).sum())
print("Fee price:", fee_price)
```

```python
print("Interest without trading fee:", sum(interest_list))
print("Interest with fee included:", sum(interest_list) - fee_price)
```
